```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("medicaldataset.csv")

df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Result', data=df)
plt.title("Heart Attack Result Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Heart rate'], bins=20, kde=True)
plt.title("Heart Rate Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(df['Systolic blood pressure'], kde=True)
plt.title("Systolic BP")

plt.subplot(1,2,2)
sns.histplot(df['Diastolic blood pressure'], kde=True)
plt.title("Diastolic BP")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Blood sugar'], bins=20, kde=True)
plt.title("Blood Sugar Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['CK-MB'], kde=True)
plt.title("CK-MB Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Troponin'], kde=True)
plt.title("Troponin Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(data=df[['Age','Heart rate','Systolic blood pressure',
                     'Diastolic blood pressure','Blood sugar','CK-MB','Troponin']])

plt.title("Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            linewidths=0.5)

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df, hue='Result')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Result', y='Age', data=df)
plt.title("Age vs Heart Attack Result")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Result', y='Blood sugar', data=df)
plt.title("Blood Sugar vs Result")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Result', y='Heart rate', data=df)
plt.title("Heart Rate vs Result")
plt.show()
```
